# PATCHES ARE ALL YOU NEED?

# 1. Introduction

- For many years, CNNs are dominant in a field of deep learning

- However, architectures based on Transformers, including VIT have compelled CNN models

- Applying Transformers to images require a lot of computation cost, especially due to the self-attention layers

- Therefore, the author proposes a novel algorithm 'ConvMixer', which is similar to MLP-Mixer:
  - Operates on patches
  - No downsampling
  - Seperates "channel-wise mixing" from "spatial mixing" of information


# 2. Model Description

- Consists of a patch embedding layer followed by repeated applications of a simple fully-convolutional block

- Consists of depthwise convolution + pointwise convolution
  - Depthwise -> Spatial locations
    - Previous studies proved depthwise could make it
  - Pointwise -> Channel locations

- BatchNormalization, and Global Pooling were also involved
